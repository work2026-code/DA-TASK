```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("orders.csv")
```

```python
import json

with open("users.json", "r", encoding="utf-8") as f:
    users_data = json.load(f)

# Handle both flat & nested JSON safely
if isinstance(users_data, dict):
    users = pd.DataFrame(users_data[list(users_data.keys())[0]])
else:
    users = pd.DataFrame(users_data)
```

```python
conn = sqlite3.connect("restaurants.db")
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
```

```python
orders.columns = orders.columns.astype(str).str.strip().str.lower()
users.columns = users.columns.astype(str).str.strip().str.lower()
restaurants.columns = restaurants.columns.astype(str).str.strip().str.lower()
```

```python
orders["user_id"] = orders["user_id"].astype(str)
users["user_id"] = users["user_id"].astype(str)

orders["restaurant_id"] = orders["restaurant_id"].astype(str)
restaurants["restaurant_id"] = restaurants["restaurant_id"].astype(str)
```

```python
orders_users = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)

final_df = pd.merge(
    orders_users,
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")

print("MCQ 1 City with highest total revenue from Gold members")
gold_city = (
    df[df["membership_type"] == "Gold"]
    .groupby("city")["order_amount"]
    .sum()
    .sort_values(ascending=False)
)
print(gold_city, "\n")
print("Answer:", gold_city.idxmax())
print("-"*60)

print("MCQ 2 Cuisine with highest average order value")
cuisine_aov = df.groupby("cuisine")["order_amount"].mean().sort_values(ascending=False)
print(cuisine_aov, "\n")
print("Answer:", cuisine_aov.idxmax())
print("-"*60)

print("MCQ 3 Users with total spend > 1000")
user_spend = df.groupby("user_id")["order_amount"].sum()
count_users = user_spend[user_spend > 1000].count()
print("Count:", count_users)
print("-"*60)

print("MCQ 4 Restaurant rating range with highest revenue")
df["rating_range"] = pd.cut(
    df["restaurant_rating"],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=["3.03.5", "3.64.0", "4.14.5", "4.65.0"]
)
rating_rev = df.groupby("rating_range")["order_amount"].sum().sort_values(ascending=False)
print(rating_rev, "\n")
print("Answer:", rating_rev.idxmax())
print("-"*60)

print("MCQ 5 Gold members: city with highest average order value")
gold_city_aov = (
    df[df["membership_type"] == "Gold"]
    .groupby("city")["order_amount"]
    .mean()
    .sort_values(ascending=False)
)
print(gold_city_aov, "\n")
print("Answer:", gold_city_aov.idxmax())
print("-"*60)

print("MCQ 6 Cuisine with lowest restaurants but high revenue")
cuisine_stats = df.groupby("cuisine").agg(
    restaurants=("restaurant_id", "nunique"),
    revenue=("order_amount", "sum")
).sort_values(["restaurants", "revenue"], ascending=[True, False])
print(cuisine_stats)
print("-"*60)

print("MCQ 7 Percentage of orders by Gold members")
gold_pct = round(
    (df[df["membership_type"] == "Gold"].shape[0] / df.shape[0]) * 100
)
print("Answer:", gold_pct, "%")
print("-"*60)

print("MCQ 8 Restaurant with highest AOV (<20 orders)")
rest_stats = df.groupby("restaurant_name").agg(
    orders=("order_id", "count"),
    avg_value=("order_amount", "mean")
)
filtered = rest_stats[rest_stats["orders"] < 20].sort_values("avg_value", ascending=False)
print(filtered.head(1))
print("-"*60)

print("MCQ 9 Highest revenue combination (Membership + Cuisine)")
combo = df.groupby(["membership_type", "cuisine"])["order_amount"].sum().sort_values(ascending=False)
print(combo)
print("Answer:", combo.idxmax())
print("-"*60)

print("MCQ  Quarter with highest revenue")
df["order_date"] = pd.to_datetime(df["order_date"])
df["quarter"] = df["order_date"].dt.to_period("Q")
quarter_rev = df.groupby("quarter")["order_amount"].sum().sort_values(ascending=False)
print(quarter_rev)
print("Answer:", quarter_rev.idxmax())
```
